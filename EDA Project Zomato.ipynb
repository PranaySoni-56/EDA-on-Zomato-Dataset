```python
df_country=pd.read_excel("Country-Code.xlsx")
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df,df_country, on='Country Code', how='left')
final_df.head()
```

```python
final_df.shape
```

```python
final_df.isnull().sum()
```

```python
final_df.columns
```

```python
Country_names=final_df.Country.value_counts().index
```

```python
Country_values=final_df.Country.value_counts().values
```

```python
#Pie Chart-Top 3 Countries
plt.pie(Country_values[:3],labels=Country_names[:3],autopct='%1.2f%%')
```

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```

```python
sns.set_theme(style='dark')
plt.figure(figsize=(12,6))
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',data=ratings,palette=['white','red','orange','yellow','green','darkgreen'])
```

Obervation:
1. Not rated is very high count
2. Maximum Rating between 2.5 and 3.4

```python
sns.countplot(x="Rating color",data=ratings,palette=['white','red','orange','yellow','green','darkgreen'])
```
